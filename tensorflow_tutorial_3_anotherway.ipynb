```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import datasets, layers, models
import numpy as np
import matplotlib.pylab as plt
import os

os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"
```

```python
# load some sample dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
# Normalize data
```

```python
x_train = x_train.reshape(-1, 28*28).astype("float32") / 255.0
print(np.shape(x_train))

x_test = x_test.reshape(-1, 28 * 28).astype("float32") / 255.0
print(np.shape(x_test))
```

```python
# input layer 0, - 28 * 28
# inner layer 1, - 512 nodes
# inner layer 2, - 256 nodes
# o/p   layer 3, - 10 nodes, note that activation function is softmax, that is
# done in compile layer by setting from_logits
model = keras.Sequential()

model.add(keras.Input(shape=(784,)))
model.summary()

model.add(layers.Dense(512, activation='relu'))
model.summary()


model.add(layers.Dense(256, activation='relu'))
model.summary()

model.add(layers.Dense(10, activation='softmax'))
model.summary()
```

```python
# Functional is more flexible than sequential, why?
# Functional input vs sequential
input = keras.Input(shape=(784,))
x = layers.Dense(512, activation='relu')(input)
x = layers.Dense(256, activation='relu')(x)
output = layers.Dense(10, activation='softmax')(x)

# x = layers.Dense(512, activation='relu')(input)
# what does this mean?
# explaination
# this is python shorthand notation for-
# dense_layer = layers.Dense(512, activation='relu')
# x = dense_layer(x)

input_temp = keras.Input(shape=(784,))
print("A:", type(input_temp))

dense_layer = layers.Dense(512, activation='relu')
print("B", type(dense_layer))
x_1 = dense_layer(input_temp)
print("C", x_1)
dense_layer = layers.Dense(10, activation='softmax')
x_2 = dense_layer(x_1)
print("D:", type(x_2))

# Finally to model
model_2 = models.Model(inputs=input, outputs=output)
print("E", model_2)
```

```python
model.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    metrics=["accuracy"],
)

model.summary()

model.fit(x_train, y_train, batch_size=32, epochs=5, verbose=2)
```
