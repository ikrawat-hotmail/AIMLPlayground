```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import datasets, layers, models
import numpy as np
import matplotlib.pylab as plt
import os

os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"
```

```python
# load some sample dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
# Note we do not need to flatten the 28*28 image
# x_train = x_train.reshape(-1, 28*28).astype("float32") / 255.0
# x_test = x_test.reshape(-1, 28 * 28).astype("float32") / 255.0

# So the way input is given is a temporal input,
# here each column of 28 pixel is given - 28 times. Feeding one column as an
# input to network
```

```python
x_train = x_train.astype("float32") / 255.0
x_test  = x_test.astype("float32") / 255.0
```

```python
# Input is not a flattened

model = keras.Sequential()
model.add(keras.Input(shape=(28, 28)))
model.add(layers.SimpleRNN(512, activation='relu', return_sequences=False, name="Intermediate_1"))
# if we wish to stack RNN layers, we need to add one more layer as below, and can set return_sequences as true
# model.add(layers.SimpleRNN(512, activation='relu', return_sequences=False, name="Intermediate_2"))
model.add(layers.Dense(10, activation='softmax', name="Output"))
model.summary()
```

```python
model.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    metrics=["accuracy"],
)

model.fit(x_train, y_train, batch_size=32, epochs=5, verbose=2)
```
